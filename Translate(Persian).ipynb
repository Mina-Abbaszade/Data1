```python
from qiskit import QuantumCircuit, Aer, execute
```

```python
qasm_sim = Aer.get_backend('qasm_simulator')
```

```python
print("1. Sara Bob ra mibinad")
qc = QuantumCircuit(7,6)
qc.rx(0.0,0)
qc.rx(0.2,1)
qc.h(2)
qc.h(3)
qc.h(4)
qc.h(5)
qc.h(6)

qc.rz(0.002,0)
qc.rz(0.202,1)
qc.crz(0.0,2,3)
qc.crz(0.012,4,5)

qc.rx(0.004, 0)
qc.rx(0.204,1)
qc.h(2)
qc.h(3)
qc.h(4)
qc.crz(0.014,5,6)

qc.crz(0.0,2,3)
qc.h(5)
qc.h(6)

qc.cx(1,2)
qc.crz(0.016,4,5)

qc.measure(2,1)
qc.barrier()

qc.h(1)
qc.cx(3,4)
qc.crz(0.018,5,6)
qc.barrier()

qc.measure(1,0)
qc.measure(4,3)
qc.barrier()

qc.cx(0,5)

qc.h(3)
qc.measure(3,2)
qc.measure(5,5)
qc.barrier()

qc.h(0)
qc.measure(0,4)

result = execute(qc, qasm_sim).result()
counts = result .get_counts()
qc.draw(output='mpl', scale=0.9)
```
